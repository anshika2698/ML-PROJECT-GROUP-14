# Essay Word Count

```python
# Convert 'essay' column into number of words
merged_df['essay_word_count'] = merged_df['essay'].apply(lambda x: len(str(x).split()))
```

```python
merged_df.tail()
```

# Grouping the the donations made per project and creating this new feature called num_donation which donates the number of donations made to a project

```python
# Step 1: Create num_donations by grouping by projectid
donation_counts = merged_df.groupby('projectid')['donationid'].count().reset_index()
donation_counts.rename(columns={'donationid': 'num_donations'}, inplace=True)
```

```python
merged_df = merged_df.merge(donation_counts, on='projectid', how='left')
```

```python
merged_df.head()
```

```python
project3_df = merged_df.groupby('projectid').agg({
    'fully_funded_bin_x': 'first',  # Target variable
    'donationid': 'count',
    'essay_word_count':'first'          # Simple feature: number of donations
}).rename(columns={'donationid': 'num_donations'})
```

```python
project3_df.head(10)
```

```python
# Step 2: Create feature matrix X and target vector y
X = project3_df[['num_donations','essay_word_count']]      # Feature(s)
y = project3_df['fully_funded_bin_x']  
```

```python
print(X)
```

# Baseline Model

# Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Baseline model
clf = LogisticRegression(class_weight='balanced')
clf.fit(X_train, y_train)

# Predict and evaluate
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Train Random Forest
rf = RandomForestClassifier(class_weight='balanced', random_state=42)
rf.fit(X_train, y_train)

# Predict and evaluate
y_pred = rf.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report

# Train XGBoost
xgb = XGBClassifier(
    objective='binary:logistic',
    
    scale_pos_weight=(y == 0).sum() / (y == 1).sum(),  # handle class imbalance
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42
)
xgb.fit(X_train, y_train)

# Predict and evaluate
y_pred = xgb.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report

# Train Gradient Boosting
gb = GradientBoostingClassifier(
    random_state=42
)
gb.fit(X_train, y_train)

# Predict and evaluate
y_pred = gb.predict(X_test)
print(classification_report(y_test, y_pred))
```
